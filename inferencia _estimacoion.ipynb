```python
import numpy as np 
import scipy.stats as scs
import math
import matplotlib.pyplot as plt
np.random.seed(12345)
```

```python
muestra = np.random.rand(10000)
plt.hist(muestra,bins=int(np.sqrt(len(muestra))),density=True)
```

```python
lambdas = [1,2,5,10]
for lambdaa in lambdas:
    y = [-1/lambdaa * math.log(1-muestrita) for muestrita in muestra]
    plt.hist(y,bins=int(np.sqrt(len(y))),density=True)
```

```python
sigam2 = 1

for lambdaaa in lambdas:
    z = [np.sqrt(2*sigam2*lambdaaa * ys ) for ys in y]
    plt.hist(z,bins=int(np.sqrt(len(z))),density=True)
```
