```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency
import scipy.stats as scs
import math
```

```python
# Cargar el archivo CSV
file_path = '/home/ig/Desktop/prueba/dataset.csv'  # Reemplaza con la ruta correcta a tu archivo
df = pd.read_csv(file_path)

# Crear listas separadas para cada clase
clase_1 = [(row['ancho'], row['largo']) for _, row in df[df['clase'] == 1].iterrows()]
clase_2 = [(row['ancho'], row['largo']) for _, row in df[df['clase'] == 2].iterrows()]

def remover_nan_tuplas(lista_de_tuplas):
    return [(ancho, largo) for ancho, largo in lista_de_tuplas if not (math.isnan(ancho) or math.isnan(largo))]

clase_1 = remover_nan_tuplas(clase_1)
clase_2 = remover_nan_tuplas(clase_2)
clase_2 = clase_2[0:486]
print(clase_2)
print(clase_1)
```

Actividad 1

```python
# Obtener solo el segundo elemento de cada tupla
anchos1 = [tupla[0] for tupla in clase_1]
anchos2 = [tupla[0] for tupla in clase_2]
plt.figure(figsize=(10,6))
# plt.hist(anchos1)#bins=int(np.sqrt(len(anchos1))))
Hx= plt.hist(anchos1,bins=7)
plt.figure(figsize=(10,6))
# plt.hist(anchos2)
Hy= plt.hist(anchos2,bins=7)
```

Actividad 2

```python
# Calcular el histograma conjunto usando np.histogram2d
H, xedges, yedges = np.histogram2d(anchos1, anchos2,bins=7)

# Configurar la figura
plt.figure(figsize=(10, 8))

# Graficar el histograma conjunto usando imshow
plt.imshow(H.T, origin='lower', aspect='auto', extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.colorbar(label='Frecuencia')
plt.xlabel('Anchos 1')
plt.ylabel('Anchos 2')
plt.title('Histograma conjunto de Anchos 1 y Anchos 2')

plt.show()
```

Actividad 3

```python
hx_noramlizado = Hx[0] / sum(Hx[0])
hy_noramlizado = Hy[0] / sum(Hy[0])
matriz2D = np.outer(hx_noramlizado,hy_noramlizado)
# Configurar la figura
plt.figure(figsize=(10, 8))

# Graficar el histograma conjunto usando imshow
plt.imshow(matriz2D.T, origin='lower', aspect='auto')
plt.colorbar(label='Frecuencia')
plt.xlabel('Anchos 1')
plt.ylabel('Anchos 2')
plt.title('Histograma conjunto de Anchos 1 y Anchos 2')

plt.show()
```
